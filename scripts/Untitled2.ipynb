```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sufix = '-v3'
RAW_PATH = '../examples/logs/phylo%s/DiseasePhyloReports.tsv'%sufix
raw_data = pd.read_csv(RAW_PATH, sep = "\t|,", engine='python')
infectious_data = raw_data[raw_data['diseaseStatus'] == 'Infectious'].reset_index()
infectious_data
```

```python
name = []
for seq in infectious_data['diseaseSeq']:
    name.append(seq.split('.')[-1])
infectious_data['case'] = name 
```

```python
def jaccards(seq1,seq2):
    seq1 = set(seq1.split('.'))
    seq2 = set(seq2.split('.'))
    
    return len(seq1 & seq2)/len(seq1 | seq2)
```

```python
distance = []
N = len(infectious_data)
for i in range(N):
    tmp = []
    seq1 = infectious_data.iloc[i]['diseaseSeq']
    for j in range(0,i+1):
        if j == i:
            tmp.append(0)
        else:
            seq2 = infectious_data.iloc[j]['diseaseSeq']
            tmp.append(1-jaccards(seq1,seq2))
    print(len(tmp))
    distance.append(tmp)
            
```

```python
samelength = []
maxlen = len(distance[-1])
for d in distance:
    samelength.append(d + [0] * (maxlen-len(d)))
```

```python
dnp = np.array(samelength)
np.save("dm%s.npy"%sufix,samelength)
```

```python
import Bio.Phylo.TreeConstruction as tc
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
from Bio import Phylo
```

```python
dm = tc.DistanceMatrix(names = list(infectious_data['case']), matrix = distance)
```

```python
constructor = DistanceTreeConstructor()
njtree = constructor.nj(dm)
```

```python
plt.figure(figsize = (10,100), dpi = 100)
Phylo.draw(njtree,do_show=False,show_confidence=False)
```
