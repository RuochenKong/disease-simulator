```python
import numpy as np
import pandas as pd
```

```python
sufix = '-v3'
RAW_PATH = '../examples/logs/phylo%s/DiseasePhyloReports.tsv'%sufix
raw_data = pd.read_csv(RAW_PATH, sep = "\t|,", engine='python')
infectious_data = raw_data[raw_data['diseaseStatus'] == 'Infectious'].reset_index()
name = []
for seq in infectious_data['diseaseSeq']:
    name.append(seq.split('.')[-1])
dm = np.load('dm-v3.npy')
```

```python
import random
reported = infectious_data[infectious_data['Reported']].reset_index()
name = []
random.seed(0)
for seq in infectious_data['diseaseSeq']:
    if random.random() < 0.3:
        name.append(seq.split('.')[-1])
len(name)
```

```python
simu_seq = open('../../pibuss/333.fasta')
simu_seq_sampled = open('../../pibuss/333_sampled.fasta','w')
istaxa = True
dowrite = False
for line in simu_seq.readlines():
    if istaxa:
        if line[1:].strip() in name:
            simu_seq_sampled.write(line)
            dowrite = True
    if not istaxa:
        if dowrite:
            simu_seq_sampled.write(line)
        dowrite = False
    istaxa = not istaxa
simu_seq_sampled.close()        
simu_seq.close()
```
